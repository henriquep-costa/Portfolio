# Análise de Volume de Treinos

Projeto simples de análise de dados utilizando Python e Pandas para analisar o volume total de treinos ao longo do tempo.

```python
import pandas as pd
```

```python
planilha = pd.read_csv('planilha_de_treinos.csv')
```

```python

pd.to_datetime(planilha['data'], dayfirst=True)
planilha.info()

```

```python

planilha['volume'] = planilha['carga kg'] * planilha['reps'] * planilha['series']
planilha.head()
```

```python
volume_por_dia = planilha.groupby('data')['volume'].sum()
volume_por_dia
```

```python
volume_por_dia = planilha.groupby('data')['volume'].sum().reset_index()
planilha.pivot_table(
    values='volume',
    index='data',
    columns='treino',
    aggfunc='sum'
)
```
